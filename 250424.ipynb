```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.style.use("ggplot")
```

```python
df = pd.read_csv('cafe_1.csv')
df
```

```python
# 그래프의 크기 설절( 가로 15, 세로 6)
plt.figure(figsize=(15, 6))

# 아메리카노 판매량을 날짜별로  선 그래프로 나타냄
sns.lineplot(
                data = df,
                x = 'date',
                y = 'americano',
                marker = 'o'    # 판매량을 표시할 때, 동그라미 표시(o)
)

# 그래프의 제목 설정
plt.title('Daily Americano Sales')

# 그래프의 x축 이름 설정
plt.xlabel('Date')

# 그래프의 y축 이름 설정
plt.ylabel('Americano Sales')

# 날짜가 많아 겹쳐서 글자를 45도 돌리기
plt.xticks(rotation=45)

#그래프 내용이 겹치지 않도록 자동으로 간격 조정
plt.tight_layout()

# 완성된 그래프를 화면에 출력
plt.show()
```

```python
df.tail(2)
```

```python
# 전체 데이터프레임에서 모든 행에 대해 americano 열부터 tiramisu 열까지 출력
menu_sales = df.loc[:, 'americano':] # 티라미수는 안써도 끝까지 출력함
menu_sales.head()
```

```python
# 메뉴별로 모든 날짜의 판매 수량을 더해서 총 판매량을 계산
menu_sales = menu_sales.sum()
menu_sales
```

```python
menu_sales = menu_sales.sort_values(ascending=False) # 내림차순으로 정렬
menu_sales
```

```python
# 메뉴명 조회
menu_sales.index
```

```python
# 메뉴별 판매량 조회
menu_sales.values
```

```python
plt.figure(figsize=(10, 6))

sns.barplot(
                x = menu_sales.values,
                y = menu_sales.index
)

plt.title('Total Sales by menu')

plt.xlabel('Total Sales')

plt.ylabel('Menu')

plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(10, 7))

plt.pie(
    menu_sales,
    labels = menu_sales.index,
    autopct = '%.2f%%',
    startangle = 140
)

plt.title('Pie Chart - TotalSales Proportion by Product')

# 파이 그래프가 항상 완벽한 원 모양으로 표시되도록 해줌
plt.axis('equal')

plt.show()
```

```python
plt.figure(figsize=(10, 6))

sns.scatterplot(
                x = 'cafe_latte',
                y = 'ny_cheese_cake',
                data=df
)

plt.title('Scatter Plot - Cafe Latte vs NY Cheese Cake')

plt.xlabel('Cafe Latte Sales')

plt.ylabel('NY Cheese Cake Sales')

# 그래프의 범례를 영어로 표시
plt.legend(['Sales Data'])

plt.tight_layout()

plt.show()
```
