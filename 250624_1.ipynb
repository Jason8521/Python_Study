```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
plt.style.use("ggplot")
```

```python
df = pd.read_csv("weight_vs_mpg.csv")
df
```

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    x="weight",
    y="mpg",
    data=df,
    alpha=0.5,
    color="blue"
)

plt.xlabel("Vehicle Weight")
plt.ylabel("Fuel Efficiency (MPG)")

plt.title("Vehicle Weight vs Fuel Efficiency")

plt. grid(True)

plt.tight_layout()

plt.show()
```

```python
X = df["weight"]
X
```

```python
weight_std = X.std()
weight_std
```

```python
weight_mean = X.mean()
weight_mean
```

```python
X = (X - weight_mean) / weight_std
weight_mean
```

```python
y = df["mpg"]
y
```

```python
a = 0
b = 0
```

```python
m = len(X)
m
```

```python
epochs = 1000
```

```python
learning_rate = 0.01
```

```python
for epoch in range(epochs):
    H = a * X + b
    
    error = H - y
    
    cost = (1/m) * np.sum(error ** 2)

    print(f'Epoch {epoch}, Cost: {cost}, a: {a}, b: {b:}')

    a_grad = (1/m) * np.sum(X * (H- y))
    b_grad = (1/m) * np.sum(H - y)

    a = a - learning_rate * a_grad
    b = b - learning_rate * b_grad

    print(f'\n최적화된 선형 회귀식: y = {a}X + {b}')
```

```python
weight_input = 100

weight_input_norm = (weight_input - weight_mean) / weight_std

predicted_mpg = a * weight_input_norm + b

print(f"\n무게가 100일 때 예측된 연비(MPG): {predicted_mpg:.4f}")
```
