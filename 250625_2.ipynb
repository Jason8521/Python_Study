```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import torch
```

```python
plt.style.use("ggplot")
```

```python
df = pd.read_csv("weight_horsepower_mpg.csv")
df
```

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    x="horsepower",
    y="mpg",
    data=df,
    color="blue"
)

plt.xlabel("Horsepower")
plt.ylabel("Fuel Efficiency (MPG)")

plt.title("Horsepower vs Fuel Efficiency")

plt.grid(True)

plt.tight_layout()

plt.show()
```

```python
weight = df["weight"]
weight
```

```python
horsepower = df["horsepower"]
horsepower
```

```python
mpg = df["mpg"]
mpg
```

```python
weight_mean = weight.mean()
weight_mean
```

```python
weight_std = weight.std()
weight_std
```

```python
weight_norm = (weight - weight_mean) / weight_std
weight_norm
```

```python
horsepower_mean = horsepower.mean()
horsepower_mean
```

```python
horsepower_std = horsepower.std()
horsepower_std
```

```python
horsepower_norm = (horsepower - horsepower_mean) / horsepower_std
horsepower_norm
```

```python
X1 = weight_norm
X1
```

```python
X2 = horsepower_norm
X2
```

```python
y = df["mpg"]
y
```

```python
X1 = torch.tensor(X1, dtype=torch.float32)
X1
```

```python
X2 = torch.tensor(X2, dtype=torch.float32)
X2
```

```python
y = torch.tensor(y, dtype=torch.float32)
y
```

```python
a1 = 0.0
a2 = 0.0
b = 0.0
```

```python
a1 = torch.tensor(a1, requires_grad=True)
a1
```

```python
a2 = torch.tensor(a2, requires_grad=True)
a2
```

```python
b = torch.tensor(b, requires_grad=True)
b
```

```python
m = len(X1)
m
```

```python
epochs = 1000
```

```python
learning_rate = 0.01
```

```python
for epoch in range(epochs):
    H = a1 * X1 + a2 * X2 + b
    error = H - y
    cost = (1/m) * torch.sum(error ** 2)

    print(f"Epoch {epoch}, Cost: {cost:.4f}, a1: {a1:.4f}, a2: {a2:.4f}, b: {b:.4f}")

    cost.backward()
    
    with torch.no_grad():
        a1 -= learning_rate * a1.grad
        a2 -= learning_rate * a2.grad
        b -= learning_rate * b.grad

        a1.grad.zero_()
        a2.grad.zero_()
        b.grad.zero_()

print(f"\n최적화된 선형회귀식: mpg = {a1:.4f} * weight = {a2:.4f} * horsepower + {b:.4f}")
```

```python
weight_input = 100
horsepower_input = 120
```

```python
weight_input_norm = (weight_input - weight_mean) / weight_std
weight_input_norm
```

```python
horsepower_input_norm = (horsepower_input - horsepower_mean) / horsepower_std
horsepower_input_norm
```

```python
predicted_mpg = a1 * weight_input_norm + a2 * horsepower_input_norm + b

print(f"\nweight=100, horsepower=120일 때 예측 연비(MPG): {predicted_mpg:.4f}")
```
