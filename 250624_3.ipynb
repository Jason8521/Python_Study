```python
import pandas as pd
import numpy as np
import tensorflow as tf
```

```python
df = pd.read_csv("weight_vs_mpg.csv")
df
```

```python
X = df["weight"]
X
```

```python
weight_std = X.std()
weight_std
```

```python
weight_mean = X.mean()
weight_mean
```

```python
X = (X - weight_mean) / weight_std
weight_mean
```

```python
y = df["mpg"]
y
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1, input_shape=[1])
])
```

```python
model.compile(
    optimizer=tf.keras.optimizers.SGD(learning_rate=0.01),

    loss='mean_squared_error'
)
```

```python
history = model.fit(
    X,
    y,
    epochs=500
)
```

```python
weight, bias = model.layers[0].get_weights()

print(f'\n학습 완료된 기울기(a): {weight[0][0]:.4f}, 절편(b): {bias[0]:.4f}')
```

```python
weight_input = 100

weight_input_norm = (weight_input - weight_mean) / weight_std

weight_input_norm
```

```python
predicted_mpg = model.predict([weight_input_norm])
```

```python
print(f"\n무게가 100일 때 예측된 연비(MPG): {predicted_mpg}")
```
