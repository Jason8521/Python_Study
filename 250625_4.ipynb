```python
import pandas as pd
import numpy as np
import tensorflow as tf
```

```python
df = pd.read_csv("weight_horsepower_mpg.csv")
df
```

```python
X = df[["weight", "horsepower"]]
X
```

```python
X_std = X.std()
X_std
```

```python
X_mean = X.mean()
X_mean
```

```python
X = (X - X_mean) / X_std
X
```

```python
y = df[["mpg"]]
y
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1, input_shape=[2])
])
```

```python
model.compile(
    optimizer=tf.keras.optimizers.SGD(learning_rate=0.01),
    loss='mean_squared_error'
)
```

```python
history = model.fit(X, y, epochs=100)
```

```python
a_values = model.layers[0].weights[0].numpy().flatten()
b_values = model.layers[0].weights[1].numpy()[0]

print(f"학습된 기울기 a1 (무게에 대한 계수): {a_values[0]:.4f}")
print(f"학습된 기울기 a2 (마력에 대한 계수): {a_values[1]:.4f}")
print(f"학습된 절편 b: {b_values:.4f}")
```

```python
new_input = pd.DataFrame([[100, 200]], columns=["weight", "horsepower"])
new_input
```

```python
new_input = (new_input - X_mean) / X_std
new_input
```

```python
predicted_mpg = model.predict(new_input)

print(f"\n예측된 연비(MPG): {predicted_mpg}")
```
