```python
import pandas as pd

data = {
    '이름': ['철수', '영희', '민수', '수진'],
    '나이': [25, 30, 22, 28],
    '성별': ['남자', '여자', '남자', '여자'],
    '점수': [90, 85, 95, 88]
}

df = pd.DataFrame(data)
df
```

```python
import pandas as pd

data = {
    '손님': ['철수', '영희', '민수', '철수', '영희', '민수'],
    '메뉴': ['피자', '피자', '피자', '햄버거', '햄버거', '햄버거'],
    '개수': [2, 1, 3, 2, 2, 1]
}

import pandas as pd
df = pd.DataFrame(data)
df
```

```python
pivot_menu = df.pivot_table(
    index = '메뉴',
    values = '개수',
    aggfunc = 'sum'
)
pivot_menu
```

```python
pivot_customer = df.pivot_table(
    index = '손님',
    values = '개수',
    aggfunc = 'sum'
)
pivot_customer
```

```python
pivot_menu = df.pivot_table(
    index = '메뉴',
    values = '개수',
    aggfunc = 'mean'
)
pivot_menu
```

```python
import pandas as pd

data = {
    '성별': ['남자', '여자', '남자', '여자', '남자', '여자', '남자'],
    '메뉴': ['피자', '피자', '햄버거', '햄버거', '피자', '햄버거', '햄버거']
}

df = pd.DataFrame(data)
df
```

```python
table = pd.crosstab(df['성별'], df['메뉴'])
table
```

```python
table2 = pd.crosstab(df['메뉴'], df['성별'])
table2
```

```python
table3 = pd.crosstab(df['성별'], df['메뉴'], margins=True)
table3
```

```python
table_percent = pd.crosstab(df['성별'], df['메뉴'], normalize=True) * 100
table_percent = table_percent.round(2)
table_percent
```

```python
table_percent_all = pd.crosstab(df['성별'], df['메뉴'], margins=True, normalize='all') * 100
table_percent_all = table_percent_all.round(2)
table_percent_all
```
