```python
students = ["철수", "영희", "민수", "지수", "준호"]
scores = [85, 92, 78, 88, 90]

total = 0

sum_text = "총합 = "

for idx, score in enumerate(scores):
    total += score
    sum_text += f"{students[idx]}({score})"
    if idx < len(scores) - 1:
        sum_text += "+"

average= total / len(scores)

print(sum_text + f" ={total}점")
print(f"평균 = {total} / {len(scores)}명 = {average:.1f}점")
```

```python
correct_password = '1234'
tries = 0
while tries < 3:
    input_password = input("비밀번호 입력:")
    if input_password == correct_password:
        print("비번 일치")
        break
    else:
        tries += 1
        print(f"비번 틀림 남은기회 {3-tries}번")

print(f"tries={tries}")
if tries == 3:
    print("비번 3회 실패!!")
```

```python
def cook_ramen(water, ramen, soup):
    '''
    이 함수는 물, 라면 종류, 스프 종류를 받아서
    라면 끓이는 과정을 출력하는 함수입니다.
    '''
    print(f"{water}ml 물을 끓입니다.")
    print(f"{ramen}을 넣습니다.")
    print(f"{soup}을 넣습니다.")
    print("라면 완성")

cook_ramen(500, "신라면", "매운맛 스프")
cook_ramen(600, "진라면", "순한맛 스프")
```

```python
def boil_noodle():
    '''
    면을 삶는 과정입니다.
    삶아진 면을 결과로 돌려줍니다.
    '''
    noodle = "삶아진 면"
    return noodle

def fresh_noodle(noodle):
    print(f"{noodle} 끓임")
    message = "라면 완성"
    return message

def make_ramen(noodle):
    '''
    삶아진 면을 받아서
    스프를 넣고 끓여서 라면을 완성하는 과정입니다.
    완성 메시지를 리턴합니다.
    '''
    print(f"{noodle}에 스프를 넣고 끓입니다.")
    message = "라면 완성!"
    return message

boiled_noodle = boil_noodle()
complete_message = make_ramen(boiled_noodle)
print(complete_message)
make_ramen(fresh_noodle)
```
