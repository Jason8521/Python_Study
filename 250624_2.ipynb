```python
import pandas as pd
import numpy as np
import torch
```

```python
df = pd.read_csv("weight_vs_mpg.csv")
df
```

```python
X = df["weight"]
X
```

```python
weight_std = X.std()
weight_std
```

```python
weight_mean = X.mean()
weight_mean
```

```python
X = (X - weight_mean) / weight_std
weight_mean
```

```python
y = df["mpg"]
y
```

```python
X = torch.tensor(X, dtype=torch.float32)
y = torch.tensor(y, dtype=torch.float32)
```

```python
a = torch.tensor(0.0, requires_grad=True)
b = torch.tensor(0.0, requires_grad=True)
```

```python
m = len(X)
m
```

```python
epochs = 1000
```

```python
learning_rate = 0.01
```

```python
for epoch in range(epochs):
    H = a * X + b
    error = H - y

    cost = (1/m) * torch.sum(error ** 2)

    print(f'Epoch {epoch}, Cost: {cost:.4f}, a: {a:.4f}, b: {b:.4f}')
    
    cost.backward()

    with torch.no_grad():
        a -= learning_rate * a.grad
        b -= learning_rate * b.grad

        a.grad.zero_()
        b.grad.zero_()

print(f'\n최적화된 선형회귀식: y = {a:.4f}X + {b:.4f}')
```

```python
weight_input = 100

weight_input_norm = (weight_input - weight_mean) / weight_std

weight_input_norm = torch.tensor(weight_input_norm, dtype=torch.float32)

predicted_mpg = a * weight_input_norm + b

print(f"무게가 100일 때 예측된 연비(MPG): {predicted_mpg:.4f}")
```
