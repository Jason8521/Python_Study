```python
import os
import pandas as pd
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import load_img, img_to_array
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D
from tensorflow.keras.layers import BatchNormalization
from tensorflow.keras.layers import Activation
from tensorflow.keras.layers import MaxPool2D
from tensorflow.keras.layers import SpatialDropout2D
from tensorflow.keras.layers import GlobalAveragePooling2D
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.initializers import HeNormal
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.callbacks import ReduceLROnPlateau
from tensorflow.keras.callbacks import ModelCheckpoint
```

```python
img_dir = r'C:\ai_project01\UTKFace'

file_names = [f for f in os.listdir(img_dir) if f.endswith('.jpg')]
file_names
```

```python
age_list = [int(f.split('_')[0]) for f in file_names]
age_list
```

```python
data = pd.DataFrame({
    'filename': file_names,
    'age': age_list
})
data
```

```python
data = data.sort_values(by="age")
data
```

```python
data.reset_index(inplace=True)
data
```

```python
df_train, df_test = train_test_split(data, test_size=0.2, random_state=42)
```

```python
df_train
```

```python
df_test
```

```python
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    horizontal_flip=True,
    brightness_range=[0.8, 1.2],
    zoom_range=0.2,
    width_shift_range=0.1,
    height_shift_range=0.1,
    fill_mode='nearest'
)
```

```python
file_name = data["filename"][0]
file_name
```

```python
img_path = f'C:/ai_project01/UTKFace/{file_name}'

image = load_img(img_path, target_size=(200, 200))
image
```

```python
image = img_to_array(image)
image
```

```python
image.shape
```

```python
image = np. expand_dims(image, axis=0)
image.shape
```

```python
aug_iter = train_datagen.flow(image, batch_size=1)
plt.figure(figsize=(12, 12))

for i in range(16):
    plt.subplot(4, 4, i+1)
    aug_image = next(aug_iter)[0]
    plt.imshow(aug_image)
    plt.axis('off')

plt.tight_layout()
```

```python
test_datagen = ImageDataGenerator(rescale=1./255)
```

```python
batch_size = 32
image_size = (200, 200)
```

```python
train_generator = train_datagen.flow_from_dataframe(
    df_train,
    directory=img_dir,
    x_col='filename',
    y_col='age',
    target_size=image_size,
    batch_size=batch_size,
    class_mode='raw',
    shuffle=True
)
```

```python
test_generator = test_datagen.flow_from_dataframe(
    df_test,
    directory=img_dir,
    x_col='filename',
    y_col='age',
    target_size=image_size,
    batch_size=batch_size,
    class_mode='raw',
    shuffle=False
)
```

```python
X_batch, y_batch = next(train_generator)

print("이미지 배치 형태:", X_batch.shape)
print("나이 배치 형태:", y_batch.shape)
print("첫번째 이미지의 나이:", y_batch[0])
```

```python
plt.figure(figsize=(16, 8))

for i in range(32):
    plt.subplot(4, 8, i + 1)
    plt.imshow(X_batch[i])
    plt.title(f'Age: {int(y_batch[i])}', fontsize=10)
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
age_model = Sequential([
    Conv2D(64, kernel_size=3, padding='same', input_shape=(200, 200, 3), kernel_initializer=HeNormal()),
    BatchNormalization(),
    Activation('relu'),
    MaxPool2D(pool_size=3, strides=2),
    SpatialDropout2D(0.1),

    Conv2D(128, kernel_size=3, padding='same', kernel_initializer=HeNormal()),
    BatchNormalization(),
    Activation('relu'),
    MaxPool2D(pool_size=3, strides=2),
    SpatialDropout2D(0.1),

    Conv2D(256, kernel_size=3, padding='same', kernel_initializer=HeNormal()),
    BatchNormalization(),
    Activation('relu'),
    MaxPool2D(pool_size=3, strides=2),
    SpatialDropout2D(0.1),

    Conv2D(512, kernel_size=3, padding='same', kernel_initializer=HeNormal()),
    BatchNormalization(),
    Activation('relu'),
    MaxPool2D(pool_size=3, strides=2),
    SpatialDropout2D(0.1),

    GlobalAveragePooling2D(),
    Dense(128, kernel_initializer=HeNormal()),
    BatchNormalization(),
    Activation('relu'),
    Dropout(0.25),
    Dense(1, activation='linear', name='age')
])
```

```python
age_model.summary()
```

```python
optimizer = Adam(learning_rate=0.01)
```

```python
callbacks = [
    EarlyStopping(monitor='val_loss', patience=30, restore_best_weights=True, verbose=1),
    ReduceLROnPlateau(monitor='val_loss', factor=0.9, patience=10, verbose=1, min_lr=1e-6),
    ModelCheckpoint('best_age_model_final.h5', monitor='val_loss', save_best_only=True, verbose=1)
]
```

```python
age_model.compile(
    optimizer=optimizer,
    loss='mse',
    metrics=['mae']
)
```

```python
history = age_model.fit(
    train_generator,
    epochs=20,
    validation_data=test_generator,
    callbacks=callbacks
)
```
