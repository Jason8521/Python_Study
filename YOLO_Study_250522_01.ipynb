```python
# ultralytics 패키지를 불러옵니다
import ultralytics

# ultralytics.checks() 함수는 현재 YOLO를 설치할 때 필요한 환경이 잘 준비되어 있는지 확인하고,
# 설치된 YOLO 버전과 사용 가능한 하드웨어(GPU, CPU 등) 정보도 보여줍니다
# 이 함수는 별도의 매개변수가 필요하지 않습니다
ultralytics.checks()

# YOLO를 사용하기 위해 ultralytics에서 YOLO 클래스를 불러옵니다
from ultralytics import YOLO

# YOLO 클래스의 객체(모델)를 생성합니다
# YOLO('모델 파일 이름') 형태로 입력하면, 미리 훈련된 모델을 불러와 바로 사용할 수 있습니다
# 여기서 'yolov8n.pt'는 YOLOv8 Nano 버전의 미리 학습된 모델로, 가장 작고 빠르며 가벼운 모델입니다
# 모델 이름 뒤어 .pt 확장자는 파이토치에서 사용하는 모델 파일 형식입니다
model = YOLO('yolov8n.pt')
```

```python
# YAML(yaml)은 설정 파일을 만들고 읽을 때 자주 사용하는 형식으로,
# 사람이 읽고 이햏하기 쉽게 만들어진 데이터 저장 방식입니다
import yaml
```

```python
# YOLO 학습에 필요한 환경 설정을 저장하는 딕셔너리(여러 정보를 한곳에 저장한 데이터 형태)를 만듭니다
yolo_config = {
    # YOLO 학습에 사용할 이미지들이 저장된 폴더의 경로입니다
    # train은 모델을 학습시킬 때 실제로 사용되는 이지미 데이터를 의미합니다
    "train": "C:/ai_project01/incredible_dataset/train/images",

    # 모델 학습 과정 성능이 얼마나 잘 나오는지 평가하기 위해 사용할 테스트 이미지의 경로입니다
    # test 데이터는 학습에 쓰이지 않고 중간에 점검한 하는 용도로 사용됩니다
    "test": "C:/ai_project01/incredible_dataset/train/images",

    # 모델 학습이 끝난 후 최종적으로 모델이 얼마나 정확한지 평가할 때 사용할 이미지 데이터의 경로입니다
    # val(validation, 검즘) 데이터라고도 하며, 역시 학습 과정에서 직접 사용하지 않습니다
    "val": "C:/ai_project01/incredible_dataset/train/images",
    
    # nc는 Number of Classes(클래스의 개수)의 약자루, 모델이 구분할 수 있는 대상(클래스)의 개수를 의미합니다
    # 현재 설정에서는 등장인물(Mr_incredible, elastic_girl, dash, violet) 4가지로 설정되어 있습니다
    "nc": 4,

    # names는 클래스의 이름을 나타내는 리스트(순서대로 나열된 데이터 구조)입니다
    # 클래스의 순서가 매우 중요하며, 이 순서대로 YOLO가 인식한 결과가 출력됩니다
    # 현재 셜정된 순서는 ["Mr_incredible", "elastic_girl", "dash", "violet"]이며,
    # 이 순서는 cvat에서 레이블 생성시 설정한 순서입니다
    "names": [
        "mr_incredible",
        "elastic_girl",
        "dash",
        "violet"
    ]
}
```

```python
# YAML 파일을 생성하고 테이터를 저장히기 위한 코드입니다

# with open()은 파일을 열 때 사용하는 파이썬의 특별한 명령입니다
# open('파일이름', '모드') 형태로 사용하며,
# '파일이름'은 생성하거나 읽고자 하는 파일의 이름이며,
# '모드'는 파일을 열 때의 용도입니다. 대표적인 모드는 아래와 같습니다.
# 'r'은 읽기(read), 'w'는 쓰기(write), 'a'는 추가(append)입니다
# 현재는 데이터를 저장하기 위해 'w'(쓰기) 모드로 파일을 엽니다

# 파일을 with 구문으로 열면, 별도로 닫지 않아도 자동으로 파일을 닫아 줍니다
# as file은 열려진 파일을 file이라는 이름으로 다루겠다는 의미입니다
with open('incredible_yolo.yaml', 'w') as file:
    
    # yaml.dump() 함수는 파이썬 데이터를 YAML 파일 형식으로 변화하여 파일에 저장하는 함수입니다
    # 매개변수 설명:
    #  - 첫 번째 매개변수(yolo_config)는 파일에 저장할 데이터압니다
    #  - 두 번째 매개변수(file)은 데이터를 저장할 파일의 객체(위에서 open으로 연 파일)입니다
    yaml.dump(yolo_config, file)
```

```python
# YOLO라는 인공지능 모델을 훈련(학습)시키는 코드입니다
# 이 모델은 '인크레더블' 애니메이션 등장인물의 생김새와 특징을 통해 학습합니다
model.train(
    # data는 YOLO가 학습할 때 어떤 데이터를 사용할지 설정하는 파일의 위치입니다
    # 파일 안에는 사진들이 있는 폴더 위치와 찾아야 하는 등장인물 목록 등이 적혀있습니다
    data="c:/ai_project01/incredible_yolo.yaml",

    # epochs(에폭)은 혼련 횟수입니다
    # 여기서는 전체 사진을 총 1000번 반복해서 보여주며 배우게 합니다
    # 사람을 ㅗ비유하면 책을 천 번 반복해 읽으며 공부하는 것과 비슷합니다
    epochs=1000,

    # imgsz(이미지 크기)는 모델으 볼 사진의 크기입니다
    # 사진을 가로, 세로 각각 640픽셀로 크기를 맞춰서 훈련합니다
    imgsz=640,

    # device는 모델을 훈련할 때 사용할 장치입니다
    # 0으로 설정하면 컴퓨터에 있는 첫 번째 GPU를 이용합니다
    # GPU는 그래칙 처리 장치로, 모델 훈련 속도를 빠르게 도와줍니다
    # 만약 GPU가 없다면 device="cpu"로 설정할 수 있습니다
    device=0,

    # patience는 훈련 도중 더 이상 좋아지지 않으면 자동으로 멈추는 기능입니다
    # '성능이 좋아진다'는 뜻은 인공지능이 사진 속의 인물을 점점 더 잘 찾아낸다는 뜻입니다
    # 여기서 patience=10은 10번 훈련을 반복했는데도 더 이상 결과가 좋아지지 않는다면,
    # 학습이 충분히 되었다고 보고 자동으로 훈련을 중단하는 것입니다
    patience=10
)
```
