```python
import cv2
import matplotlib.pyplot as plt
import dlib
import numpy as np
from tensorflow.keras.models import load_model
```

```python
cap = cv2.VideoCapture(0)
```

```python
detector = dlib.get_frontal_face_detector()
```

```python
age_model = load_model('best_age_model_final.h5')
```

```python
predictor = dlib.shape_predictor("shape_predictor_68_face_landmarks.dat")
```

```python
while cap.isOpened() == True:
    success, image = cap.read()

    if success == False:
        continue

    faces = detector(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))

    for face in faces:
        print("얼굴 찾음!!!")
        print("얼굴 위치=", face)
        print("="*100)

        rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        x1, y1, x2, y2 = face.left(), face.top(), face.right(), face.bottom()

        x1_new = max(0, x1 - 50)
        y1_new = max(0, y1 - 50)
        x2_new = min(rgb_image.shape[1], x2 + 50)
        y2_new = min(rgb_image.shape[0], y2 + 50)

        face_img = rgb_image[y1_new:y2_new, x1_new:x2_new]

        if face_img.size == 0:
            continue

        face_img = cv2.resize(face_img, (200, 200))
        face_img = face_img / 255.0
        face_input = np.expand_dims(face_img, axis=0)
        predicted_age = age_model.predict(face_input)[0][0]
        cv2.putText(image, f'Age: {int(predicted_age)}', (x1, y1-10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 0), 2)

        #cv2.rectangle(image, (face.left(), face.top()), (face.right(), face.bottom()), (0, 255, 0), 2)
        cv2.rectangle(image, (x1_new, y1_new), (x2_new, y2_new), (0, 255, 0), 2)
        landmarks = predictor(image, face)

        for n in range(0, 68):
            x = landmarks.part(n).x
            y = landmarks.part(n).y
            cv2.circle(image, (x, y), 2, (0, 0, 255), -1)

    cv2.imshow('webcam_window01', image)

    key = cv2.waitKey(1)

    if key == ord('q'):
        break
```

```python
cap.release()
cv2.destroyAllWindows()
```
